#  Case study on Presentation of data


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Read the dataset to python environment.

```python
df= pd.read_csv('G:\my trials\cars_data.csv')
df
```

##### Check for the null values present in the dataset

```python
print(df.isnull().sum())
```

##### Plot a bar graph of male vs female buyers participated in the sales

```python
Gender= df['Buyer Gender'].value_counts().to_frame().reset_index()
for i,j in zip(Gender['index'],round(Gender['Buyer Gender'],2)):
 plt.annotate(str(j),(i,j+2))
plt.bar(Gender['index'],Gender['Buyer Gender'],color = ['red','blue'])
plt.title(' Male vs Female buyers participated in the sales')
plt.xlabel('Gender')
plt.ylabel('Count') 
Gender
```

##### Find the top 5 cars based on their sales price

```python
top_car=df.nlargest(5, ['Sale Price'])
top_car
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_car['Model'],top_car['Sale Price'],color = ['red','blue','green','orange','yellow'])
for i,j in zip(top_car['Model'],round(top_car['Sale Price'],2)):
 plt.annotate(str(j),(i,j+3))
plt.title(' Top 5 car model participated in the sales')
plt.xlabel('Model')
plt.ylabel('Sale Price') 
```

##### Find the least 5 cars based on their Resell price

```python
least_car=df.nsmallest(5, ['Resell Price'])
least_car
```

```python
plt.figure(figsize=(8,5))
plt.bar(least_car['Model'],least_car['Resell Price'],color = ['purple','violet','indigo','pink','grey'])
for i,j in zip(least_car['Model'],round(least_car['Resell Price'],2)):
 plt.annotate(str(j),(i,j+3))
plt.title(' Least 5 car model participated in the sales')
plt.xlabel('Model')
plt.ylabel('Resell Price')
```
